```python
# rabbit leap problem

# rules
    # 1. EW -> WE
    # 2. EE -> E_E
    # 3. _W -> W_
    # 4. E_ -> _E
    # 5. EW_ -> _WE
    # rule 3 is applied first so first state will be EEEW_WW


from collections import deque

def get_successors(state):
    successors = []
    empty = state.index('_')
    n = len(state)

    # W moves left
    if empty < n - 1 and state[empty + 1] == 'W':  # step left into _
        s = list(state)
        s[empty], s[empty + 1] = s[empty + 1], s[empty] # rule  3. _W -> W_
        successors.append(tuple(s))
    if empty < n - 2 and state[empty + 2] == 'W' and state[empty + 1] == 'E':  # jump left over E
        s = list(state)
        s[empty], s[empty + 2] = s[empty + 2], s[empty]  # rule 1. EW -> WE
        successors.append(tuple(s))

    # E moves right
    if empty > 0 and state[empty - 1] == 'E':  # step right into _
        s = list(state)
        s[empty], s[empty - 1] = s[empty - 1], s[empty] # Rule 4. E_ -> _E
        successors.append(tuple(s))
    if empty > 1 and state[empty - 2] == 'E' and state[empty - 1] == 'W':  # jump right over W
        s = list(state)
        s[empty], s[empty - 2] = s[empty - 2], s[empty] # rule 5. EW_ -> _WE
        successors.append(tuple(s))
    if empty>1 and state[empty-1]=='E' and state[empty-2]=='E':
        s=list(state)
        s[empty - 1],s[empty]=s[empty],s[empty - 1] # rule 4. E_ -> _E
        successors.append(tuple(s))

    return successors

def bfs(start, goal):
    q = deque([(start, [])])
    visited = set()
    while q:
        state, path = q.popleft()
        if state in visited:
            continue
        visited.add(state)
        new_path = path + [state]
        if state == goal:
            return new_path
        for nxt in get_successors(state):
            q.append((nxt, new_path))
    return None

start = ('E','E','E','_','W','W','W')
goal  = ('W','W','W','_','E','E','E')

solution = bfs(start, goal)

if solution:
    print("Solution Exist")
    for i, step in enumerate(solution):
        print(f"Step {i} : {step}")
    print(f"Total steps: {len(solution)-1}")
else:
    print("No solution")
```
